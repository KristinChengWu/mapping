```python
from scanning import camera, coordinates, visualization, optimization
```

```python
# Make a TelescopePattern
prime_cam = camera.PrimeCam()
prime_cam.add_module('Mod280', 'i1', mod_rot=0)
print(prime_cam)

# Make a SkyPattern
#sky_pattern = coordinates.Daisy('Daisy.json')
sky_pattern = coordinates.SkyPattern('Pong.csv')
visualization.sky_path(sky_pattern)

telescope_pattern = coordinates.TelescopePattern(
    sky_pattern, instrument=prime_cam, module='c', 
    obs_param='Observation.json'
)
```

```python
visualization.sky_path(telescope_pattern)
```

```python
s = optimization.Simulation(telescope_pattern, 'Mod280', max_acc=0.5, pxan_list=[(0, 0)], wafers_on=[0], )
```

```python
visualization.hitmap(s, convolve=True)
```

```python
visualization.hitmap(s, convolve=True)
```

```python
visualization.pxan_det(s, norm_pxan=True)
```
